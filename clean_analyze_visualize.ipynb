```python
less_than_average=[]
greater_than_average=[]
for values in energy['%Renewable']:
    if values<50:
        less_than_average.append(values)
    elif values>50:
        greater_than_average.append(values)
a=(len(less_than_average))
b=(len(greater_than_average))
print(len(less_than_average))

print(len(greater_than_average))
percentage_below_average = (a*100)/227
percentage_above_average = (b*100)/227

print(percentage_below_average)
print(percentage_above_average)
```
